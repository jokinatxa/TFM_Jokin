## Filtrar Informacion de partidos

Hay que filtrar la informacion sobre los partidos que se han descargado para solo tener informacion sobre las temporadas en los que el jugador haya jugado en los equipos de primera division.

```python
!unzip datos_tablas.zip
```

```python
import os
import zipfile

directorio_archivos = 'datos_tablas/datos_tablas/'
archivo_nombres_anios = 'JugadoresAño.txt'

# Leer nombres y años desde el archivo
with open(archivo_nombres_anios, 'r') as file:
    nombres_y_años = [line.strip() for line in file]

# Obtener la lista de archivos en el directorio
archivos_en_directorio = os.listdir(directorio_archivos)

# Filtrar los archivos que coinciden con la lista
archivos_seleccionados = []

for archivo in archivos_en_directorio:
  nombre, _, año = archivo.rsplit('_', 2)
  año = año.replace('.csv', '')
  nombre_año = nombre+"_"+año
  if nombre_año in nombres_y_años:
    archivos_seleccionados.append(archivo)


# Crear un archivo ZIP para almacenar los archivos seleccionados
nombre_zip = 'archivos_seleccionados.zip'
with zipfile.ZipFile(nombre_zip, 'w') as zip_file:

    for archivo_seleccionado in archivos_seleccionados:
        ruta_completa = os.path.join(directorio_archivos, archivo_seleccionado)
        zip_file.write(ruta_completa, arcname=archivo_seleccionado)

print(f'Archivos seleccionados comprimidos en {nombre_zip}')
```

## Crear df con estos archivos seleccionados.

```python
!unzip archivos_seleccionados.zip -d partidos
```

```python
import os
import pandas as pd

path_to_csv = "partidos/"
dfs = []

# Iterar sobre cada archivo en el directorio
for filename in os.listdir(path_to_csv):
    if filename.endswith(".csv"):
        file_path = os.path.join(path_to_csv, filename)

        try:
            # Leer el archivo CSV
            df = pd.read_csv(file_path)

            # Agregar una columna con el nombre del archivo
            df['NombreArchivo'] = filename

            # Agregar el DataFrame a la lista
            dfs.append(df)
        except pd.errors.ParserError as e:
            print(f"Error al procesar {filename}: {e}")
            continue

# Concat los DataFrames
df_partidos = pd.concat(dfs, ignore_index=True)

print(df_partidos)
```

Algunos archivos dan error al procesarlos, se van a ignorar esos errores, ya que son muy pocos archivos y en un analisis tan general no merece.
